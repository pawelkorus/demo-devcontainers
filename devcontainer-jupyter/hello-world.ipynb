# Elliptic curve cryptograpy

## Finite fields

### secp256r1 (NIST P-256)

Field definition:

$\mathbb{F}_{p}$, where $p = 2^{256} - 2^{224} + 2^{192} + 2^{96} - 1$

Curve equation:

$y^2 = x^3 - 3x + b$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the NIST P-256 curve
a = -3
b = int("5AC635D8AA3A93E7B3EBBD55769886BC651D06B0CC53B0F63BCE3C3E27D2604B", 16)

# Define the range for x
x = np.linspace(-1000000000000000000, 1000000000000000000, 1000)

# Calculate y^2 = x^3 + ax + b
y_squared = x**3 + a*x + b

# Filter out negative y_squared values
valid = y_squared >= 0
x_valid = x[valid]
y_valid = np.sqrt(y_squared[valid])

# Plot the positive and negative square roots of y^2
plt.figure(figsize=(8, 6))
plt.plot(x_valid, y_valid, label=r'$y = \sqrt{x^3 - 3x + b}$', color='blue')
plt.plot(x_valid, -y_valid, label=r'$y = -\sqrt{x^3 - 3x + b}$', color='red')

# Set the labels and title
plt.title(r'NIST P-256 Elliptic Curve: $y^2 = x^3 - 3x + b$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
